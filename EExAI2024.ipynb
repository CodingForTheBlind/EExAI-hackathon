```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train_dataset.csv")
dft = pd.read_csv("public_dataset_without_gt.csv")
```

```python
dft
```

```python
dfm = df.drop(columns=["default_12month"])
def create_train_test_flag(df_train, df_test):
    df_train['train_flag'] = 1
    df_test['train_flag'] = 0
    total_df = pd.concat([df_train, df_test], axis = 0,ignore_index=True)
    return total_df
#32523
dfX = create_train_test_flag(dfm, dft)
dfX = dfX.drop(columns=["ID","pms_i_ymd","date_of_birth","Province","Area","Shop Name"])
y = df["default_12month"]
```

```python
def feature_type(total_data):
  categorical_cols = total_data.select_dtypes(include=['object']).columns
  numerical_cols = total_data.select_dtypes(include=['int64', 'float64']).columns
  return categorical_cols , numerical_cols
feature_type(dfX)
```

```python
dfX = dfX.fillna(0)
sexdum = pd.get_dummies(dfX["gender"],dtype=int)
applydum = pd.get_dummies(dfX["apply"],dtype=int)
enqudum = pd.get_dummies(dfX["Leasing enquiry_count"],dtype=int)
dfX.drop(columns=["gender","apply","Leasing enquiry_count"],inplace=True)
dfX = pd.concat([dfX,sexdum,applydum,enqudum],axis=1)
```

```python
dfX
```

```python
x_test = dfX.iloc[32524:,:].copy(deep=True)
dfX = dfX.iloc[:32524,:]
Xy= pd.concat([dfX,y],axis=1)
```
